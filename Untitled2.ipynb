```python
## The FLASH image needs to be skull-stripped using BET; the MP2RAGE 2nd inv can be masked using the brain mask from fmriprep
from nipype.interfaces import fsl

def skullstrip(fn, frac=0.3, padding=True):
    if fn.endswith('.nii.gz'):
        new_fn = fn.replace('.nii.gz', '_bet.nii.gz')
    else:
        new_fn = fn.replace('.nii', '_bet.nii.gz')
    better = fsl.BET()
    better.inputs.in_file = fn
    better.inputs.out_file = new_fn
    better.inputs.mask = True
    better.inputs.padding = padding
    better.inputs.frac = frac
    better.run()
    
    # make sure affine is correct - BETting can mess this up
    hdr_old = nib.load(fn)
    hdr_new = nib.load(new_fn)
    if not (hdr_new.affine == hdr_old.affine).all():
        nib.Nifti1Image(hdr_new.get_fdata(), affine=hdr_old.affine).to_filename(new_fn)
        # do the same for mask..
    return new_fn

def apply_mask(img, mask):
    data_ = img.get_fdata()
    mask_data = mask.get_fdata() #
    data_[mask_data==0] = 0
    
    return nib.Nifti1Image(data_, header=img.header, affine=img.affine)

# def skullstrip_mp2rage():
    

def reorient_to_ras(fn):
    from nipype.interfaces.image import Reorient
    reorient = Reorient(orientation='RAS')
    reorient.inputs.in_file = fn
    res = reorient.run()
    
    return res.outputs.out_file

# shutil.copy('./sourcedata/sourcedata/ds-01/sub-01/anat/sub-01_echo-3_FLASH.nii.gz', './ds-01_sub-01_FLASH_echo_3.nii.gz')
# skullstrip('./ds-01_sub-01_FLASH_echo_3.nii.gz')
# for img in ['./ds-01_sub-01_t1w.nii.gz', './ds-01_sub-01_mean_flash.nii.gz']:
#     skullstrip(img)
```

```python
def prepare_flash_and_mp2rage(sub, working_dir='./coregister_FLASH_to_MP2RAGE'):
    os.makedirs(working_dir, exist_ok=True)
    
    print(sub + '...', end='')
    mp2rage_inv2_fn = f'./sourcedata/sourcedata/ds-01/sub-{sub}/anat/sub-{sub}_inv-2_part-mag_MP2RAGE.nii'
    t1w_mask_fn = f'./derivatives/ds-01/fmriprep/sub-{sub}/anat/sub-{sub}_desc-brain_mask.nii.gz'
    flash_echo_1_fn = f'./sourcedata/sourcedata/ds-01/sub-{sub}/anat/sub-{sub}_echo-1_FLASH.nii.gz'
    
    # BET flash and move to working dir
    print('betting...', end='')
    flash_betted = skullstrip(flash_echo_1_fn)
    shutil.move(flash_betted, os.path.join(working_dir, f'sub-{sub}_echo-1_FLASH_bet.nii.gz'))
    shutil.move(flash_betted.replace('_bet.nii.gz', '_bet_mask.nii.gz'), os.path.join(working_dir, f'sub-{sub}_echo-1_FLASH_bet_mask.nii.gz'))
    
    # apply mask to mp2rage; requires first a reorientation to RAS
    print('reorienting, masking...')
    mp2rage_inv2_ras_fn = reorient_to_ras(mp2rage_inv2_fn)
    mp2rage_inv2_masked = apply_mask(nib.load(mp2rage_inv2_ras_fn), nib.load(t1w_mask_fn))
    mp2rage_inv2_masked.to_filename(os.path.join(working_dir, f'sub-{sub}_inv-2_part-mag_MP2RAGE_masked.nii.gz'))
    
prepare_flash_and_mp2rage('01')
```

```python
def warp_flash_to_mp2rage(sub, working_dir='./coregister_FLASH_to_MP2RAGE'):
    src = os.path.join(working_dir, f'sub-{sub}_echo-1_FLASH_bet.nii.gz')
    tgt = os.path.join(working_dir, f'sub-{sub}_inv-2_part-mag_MP2RAGE_masked.nii.gz')
    
    print('Step 1: Rigid only')
    # step 1: rigid only
    output1 = registration.embedded_antsreg(source_image=src, target_image=tgt, run_rigid=True, run_affine=False, run_syn=False, save_data=True, output_dir=working_dir)

    print('Step 2: SyN only')
    # step 2: syn only, high regularization
    output2 = registration.embedded_antsreg(source_image=output1['transformed_source'], target_image=tgt, run_rigid=False, run_affine=False, run_syn=True, save_data=True, output_dir=working_dir)

    print('Combining warps for a final output...')
    registration.apply_coordinate_mappings(image=src, mapping1=output1['mapping'], mapping2=output2['mapping'], save_data=True, output_dir=working_dir, file_name=f'sub-{sub}_echo-1_FLASH_bet_final_warped.nii.gz',)

    return output1, output2

warp_flash_to_mp2rage('01')
```
